```python
def get_ascii_char(number):
    """
    Convert an integer (0-127) to its ASCII character.
    
    Args:
        number: Integer from 0 to 127
    
    Returns:
        The corresponding ASCII character as a string
    """
    if not isinstance(number, int) or number < 0 or number > 127:
        return f"Error: Number must be between 0 and 127, got {number}"
    
    return chr(number)


# Example usage
if __name__ == "__main__":
    # Test some common ASCII values
    print(f"65 -> '{get_ascii_char(65)}'")  # A
    print(f"97 -> '{get_ascii_char(97)}'")  # a
    print(f"48 -> '{get_ascii_char(48)}'")  # 0
    print(f"32 -> '{get_ascii_char(32)}'")  # space
    print(f"33 -> '{get_ascii_char(33)}'")  # !
    
    print("\nFull ASCII table (32-126):")
    for i in range(32, 127):
        print(f"{i:3d}: '{chr(i)}'", end="  ")
        if (i - 31) % 5 == 0:
            print()  # New line every 5 characters
```
